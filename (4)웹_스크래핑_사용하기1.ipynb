```python
# # 네이버에 "야구" 검색 결과 페이지 가져오기
import requests
a = '야구'    # 검색어
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={}'
r=requests.get(url.format(a))
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# 위 검색 결과의 야구 일정 가져오기
prd_link = soup.find('a', attrs={'class':'match'})
print(prd_link)
```

```python
print(prd_link['href'])
```

```python
url = 'https://search.naver.com/search.naver'+prd_link['href']
r=requests.get(url)
```
